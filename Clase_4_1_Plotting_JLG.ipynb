_Ejercicio_:regenerar X2 el número de puntos del histograma y los beans para que el histograma se parezca a una gausiana

```python
X2 = np.random.randn(10000)* 20 + 50
plt.hist(X2,bins=50, density=True)
```

Esta forma es metiendo más puntos. Pero también se podría hacer esto:

```python
def pintar_histograma_vs_teorica(numero_de_puntos):
    mu = 50
    sigma = 20
    X2 = np.random.randn(numero_de_puntos)*sigma + mu
    from scipy.stats import norm
    dist = norm(mu,sigma)
    n = 50 
    xnorm = np.linspace(mu-3*sigma, mu+3* sigma , n)
    ynorm = dist.pdf(xnorm)
    plt.plot(xnorm,ynorm, 'r-', lw=3, label='r$\mu=%s, \sigma =%s$'% (mu,sigma) )
    plt.hist(X2,bins=50,density=True, label='Histograma')
    plt.legend();
    plt.xlabel(f'Comparativa teórica vs {numero_de_puntos} aleatorios')
```

```python
pintar_histograma_vs_teorica(50)
```

_Nota_: con pocos puntos se parece poco a la teórica...

```python
pintar_histograma_vs_teorica(20000)
```

_Note_: con muchos puntos se parece muchísimo

```python
plt.plot(X,Y1,'r', label="Y1")
plt.plot(X,Y2,'g', label="Y2")
plt.fill_between(X,Y1,Y2, where=X>4)
plt.legend()
plt.xlim([-7,7])
```
