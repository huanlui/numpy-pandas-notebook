# Visualisation with Python

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Lines

```python
X = np.linspace(0,10,50)
Y = np.sin(X)

plt.plot(X,Y,'g', label=r'sin($\alpha$)')
plt.plot(X,X,'r.-',label='lin')
plt.legend(loc='best')
```

_Nota:_ la _r_ indica que es un texto espcecial, por ejenplo que acepta Latex

_Nota 2:_ recuerda que para escribir algo que es latex hay que ponerlos entre '$'

_Nota 3:_ \alpha es para escribir la letra griega

_Nota 4_:

* '.-': línea con puntos
* '--': línea discontinua
* 'o': línea discontinua
* 'k':negro (b es para blue)
* Resto: ver ayuda

_Nota 5_: loc es para decidir donde poner la legenda. Best te detecta él el sitio más adecuado


## Ejercicio

Pintar la siguiente función: $f(x)=\frac{e^x}{1+e^x}$

```python
X = np.linspace(0,10,50)
Y = np.e ** X / (1 + np.e**X)
```

```python
plt.plot(X,Y,'g', label=r'$\frac{e^x}{1+e^x}$')
plt.legend(loc='best',fontsize='xx-large')
```

Otra forma de elegir los parámetros de la figura:


```python
def set_figure_properties():
    plt.rcParams['figure.figsize'] = (10.0,8.0)
    plt.rcParams['savefig.dpi'] = 300
    plt.rcParams['figure.facecolor'] = 'white'
    sizefont = 30
    font = {'family':'serif', 'weight':'normal', 'size':sizefont}
    plt.rc('font',**font)
    plt.rc('text',usetex=False)
    plt.rc('xtick', labelsize=sizefont)
    plt.rc('ytick',labelsize=sizefont)
```

```python
set_figure_properties()
```

_Nota_ Esto se podría guardar en un mystyles.py y hacer un from mystyles import set_figure_properties

La función logística, o _sigmoide_ es muy similar a la tangente hiperbólica . Comparémoslas

```python
xlim_l = -50
xlim_r = 50
n = 500
X =np.linspace(xlim_l,xlim_r,n)
Y1 = (np.tanh(X)+1)/2
Y2 = np.exp(X) / (1.0+np.exp(X));
plt.plot(X,Y1,'b',label=r'tanh' )
plt.plot(X,Y2,'r',label=r'sig' )
plt.xlim([-5,5])
plt.legend(loc='best',fontsize='medium')
```
