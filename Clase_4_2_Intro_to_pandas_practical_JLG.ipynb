Ahora queremos traducir el DepTime, pueso como número, me lo traduzca en un string tipo HH:MM
Ejemplo: el número 1810.0 debre ser las 18:10

```python
def deptime_to_string(num):
    if num!=num: #NaN
        num = 0.0 
    hour = int(num/100) % 24
    minute = int(num%100) % 24
    return '%02d:%02d' % (hour,minute)

deptime_to_string(810.0)
```

También queremos arreglar las fechas, que están como strings

```python
ts = pd.to_datetime('2015-01-15 08:30')
ts
```

Ahora quiere que solo nos quedemos con los que tengan más de treinta vuelos

```python
mean_and_counts = df3.groupby('Origin')["DepDelay"].agg(['mean','count'])

only_more_than_30 = mean_and_counts [ mean_and_counts['count'] > 30]
only_more_than_30.sort_values('count').head()
```

```python
highly_delayed = only_more_than_30.sort_values('mean',ascending=False).head(10)
highly_delayed
```
